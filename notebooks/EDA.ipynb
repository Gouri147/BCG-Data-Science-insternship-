```python
client_df.columns
price_df.columns
```

```python
price_df['price_date'] = pd.to_datetime(price_df['price_date'])
```

```python
price_agg = price_df.groupby('id').mean().reset_index()

price_agg.head()
```

```python
merged_df = pd.merge(client_df, price_agg, on='id', how='inner')

merged_df.shape
merged_df.head()
```

```python
merged_df.columns
```

```python
merged_df.groupby('churn')[[
    'price_off_peak_var',
    'price_peak_var',
    'price_mid_peak_var',
    'price_off_peak_fix',
    'price_peak_fix',
    'price_mid_peak_fix'
]].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

price_cols = [
    'price_off_peak_var',
    'price_peak_var',
    'price_mid_peak_var',
    'price_off_peak_fix',
    'price_peak_fix',
    'price_mid_peak_fix'
]

for col in price_cols:
    plt.figure()
    sns.boxplot(x='churn', y=col, data=merged_df)
    plt.title(f"{col} vs Churn")
    plt.show()
```

```python
merged_df['avg_price'] = merged_df[[
    'price_off_peak_var',
    'price_peak_var',
    'price_mid_peak_var',
    'price_off_peak_fix',
    'price_peak_fix',
    'price_mid_peak_fix'
]].mean(axis=1)
```

```python
merged_df.groupby('churn')['avg_price'].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='churn', y='avg_price', data=merged_df)
plt.title("Average Price vs Churn")
plt.show()
```

There is no significant difference in average pricing between churned and retained customers, as both groups show similar distributions and median values. This suggests that price is not the primary driver of customer churn.

Since pricing does not strongly differentiate churn behavior, customers are likely influenced by other factors, such as:

Customer service quality
Contract terms and flexibility
Overall customer experience
